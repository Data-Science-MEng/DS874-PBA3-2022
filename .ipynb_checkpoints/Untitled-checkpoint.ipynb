```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import confusion_matrix
#Import the train_test_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
```

```python
data = pd.read_csv("heart_2020_cleaned_1.csv")
```

```python
data.head(10)
```

```python
minValuesObj = data.min()
print('minimum value in each column : ')
print(minValuesObj)
```

```python
maxValuesObj = data.max()
print('max value in each column : ')
print(maxValuesObj)
```

```python
columns_list = list(data.columns)
X = data[columns_list[1:18]]  #Input data 
y = data[columns_list[0]]   #Target data
X_cols = X
#Recode age category as numerical
encode_AgeCategory = {'55-59':57, '80 or older':80, '65-69':67,
                      '75-79':77,'40-44':42,'70-74':72,'60-64':62,
                      '50-54':52,'45-49':47,'18-24':21,'35-39':37,
                      '30-34':32,'25-29':27}

X['AgeCategory'] = X['AgeCategory'].apply(lambda x: encode_AgeCategory[x])
X['AgeCategory'] = X['AgeCategory'].astype('float')

X.info()

#Recode health category as numerical
encode_HealthCategory = {'Poor':0,'Fair':1,'Good':2,'Very good':3,'Excellent':4}

X['GenHealth'] = X['GenHealth'].apply(lambda x: encode_HealthCategory[x])
X['GenHealth'] = X['GenHealth'].astype('float')

X.info()

#One-hot encode categorical input data (INCLUDING AGECAT)
num_cols = ["BMI", "PhysicalHealth", "MentalHealth", "AgeCategory", "GenHealth", "SleepTime"]
cat_cols = columns_list[1:18]
for i in num_cols:
  cat_cols.remove(i)

X = pd.get_dummies(X, prefix=cat_cols)
X.info()

#scale input data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)


headtest = pd.DataFrame(X)
print("head before oversampling")
print(headtest.head(10))

#label encode target feature
encoder = preprocessing.LabelEncoder()
y = encoder.fit_transform(y)
print("head before oversampling")
headtest = pd.DataFrame(y)
print(headtest.head(10))


#Perform random oversampling
oversample = RandomOverSampler(sampling_strategy='minority')
X, y = oversample.fit_resample(X, y) #fit and apply the transform
print(Counter(y)) # summarize class distribution


#Create the data subsets (Train and Testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

print("The data is ready for modelling.")
```

```python
pd.DataFrame(X_train).head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier
start = datetime.now()

# Modelling
rf_model = RandomForestClassifier(n_estimators=100, )
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Accuracy metrics
from sklearn.metrics import (f1_score, precision_score,
                             recall_score, accuracy_score,
                             matthews_corrcoef)


accuracy_rf = accuracy_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf, average='weighted')
precision_rf = precision_score(y_test, y_pred_rf, zero_division=0, average='weighted')
f1s_rf = f1_score(y_test, y_pred_rf, average='weighted')
MCC_rf = matthews_corrcoef(y_test, y_pred_rf)


print("Accuracy: "+ str(round(accuracy_rf,5)))
print("Recall: "+ str(round(recall_rf,5)))
print("Precision: "+ str(round(precision_rf,5)))
print("F1-Score: "+ str(round(f1s_rf,5)))
print("MCC: "+ str(round(MCC_rf,5)))

end = datetime.now()
runtime = end - start
print("Runtime =", round(runtime.total_seconds()), " s")
print("Confusion matrix:")
confusion_matrix(y_test, y_pred_rf)
```
